```python
people = {
    "first" : ["Corey","Jane","John","Adam"],
    "last"  : ["Schafer","Doe","Doe","Doe"],
    "email" : ["Corey@gmail.com","Jane@gmail.com","John@yahoo.com","A@email.com"]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
# ***********************sorting
```

```python
# sort by a columns
df.sort_values(by = "last",ascending = False)
```

```python

# the list argument will first deal with "last" and if the last names
# are same then it will consider the second list element ie "first"
df.sort_values(by = ["last","first"],ascending = False) # list of columns
```

```python
df.sort_values(by = ["last","first"],ascending = [False,True],inplace = True)
```

```python
df
```

```python
df.sort_index(inplace = True)  # sorting by index
```
